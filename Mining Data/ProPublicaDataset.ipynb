```
response = requests.get(PROPUBLICA_SOURCE.format(99))
data = json.loads(response.text)
organization = data['organizations'][0]
```

```
response2 = requests.get(PROPUBLICA_SOURCE2.format(organization["ein"]))
data2 = json.loads(response2.text)
```

```
for i in range(0,100):
  response = requests.get(PROPUBLICA_SOURCE.format(i))
  print(i)
  data = json.loads(response.text)["organizations"]
  for obj in data:
    response2 = requests.get(PROPUBLICA_SOURCE2.format(organization["ein"]))
    data2 = json.loads(response2.text)["organization"]
    row = "{},{},{},{},{},{},{},{},{},{},{},{},{},{},{},{},{},{}\n".format(obj['ein'], obj['strein'], obj['name'], 
                                           obj['sub_name'], obj['city'], obj['state'],
                                           obj['ntee_code'], obj['score'], data2['id'],
                                           data2['careofname'], data2['address'],
                                           data2['zipcode'], data2['ruling_date'],
                                           data2['asset_amount'], data2['income_amount'],
                                           data2['revenue_amount'], data2['created_at'],
                                           data2['updated_at'])
    f.write(row)
    
```
