```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans


# parameters of the numbers of clusters
num = 2

for each_ms in mission_statement:
    # create feature vectors
    count_vect = CountVectorizer()
    each_ms_counts = count_vect.fit_transform(each_ms)
    
    # TF-IDF i.e Term Frequency times inverse document frequency.
    tfidf_transformer = TfidfTransformer()
    each_ms_tfidf = tfidf_transformer.fit_transform(each_ms_counts)
    
    # K-means clustering
    kmeans = KMeans(n_clusters=num).fit(each_ms_tfidf)
```

```python
lines_for_predicting = ["google and facebook is awesome!", 
                        "tf df increase",'what']

kmeans.predict(tfidf_transformer.fit_transform(lines_for_predicting))
```
